```python
# Measuring Time
# 1.
import time

start_time = time.time()
for i in range(1, 101):
    print(i)

print(time.time() - start_time)

"""
Problems with this approach

1. Different time for different algorithm
2. Time varies if implementation changes
3. Different machines different time
5. Does not work for extremely small input
6. Time varies for different inputs, but can't establish a relationship
"""
```

```python
import time
start_time = time.time()

j = 1
while j<101:
    print(j)
    j += 1

print(time.time() - start_time)
```
