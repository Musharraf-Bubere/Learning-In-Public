```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) Applewebkit/537.36 (KHTM, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
name = []
rating = []
reviews = []
type_location = []

for i in company:

    name.append(i.find('h2').text.strip())
    rating.append(i.find('div', class_='rating_star_container').text.strip())
    reviews.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())
    type_location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())

df=pd.DataFrame({
    'name':name,
    'rating':rating,
    'reviews':reviews,
    'type_location':type_location
   })
```

```python
final = pd.DataFrame()

for j in range(1, 10):

    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) Applewebkit/537.36 (KHTM, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')

    company = soup.find_all('div', class_='companyCardWrapper')

    name = []
    rating = []
    reviews = []
    type_location = []

    for i in company:

        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_='rating_star_container').text.strip())
        reviews.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())
        type_location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())

d=pd.DataFrame({
    'name':name,
    'rating':rating,
    'reviews':reviews,
    'type_location':type_location
    })
df = pd.DataFrame(d)

final = pd.concat([final,df], ignore_index=True)
```
